```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.inspection import permutation_importance

# File definitions
data_path = "data"
file_names = {
    "bitcoin": "bitcoin.csv",
    "gold": "gold.csv",
    "google_trends": "google_trends.csv",
    "sp500": "sp500.csv",
    "treasury_3m": "treasury_3m.csv",
    "treasury_10y": "treasury_10y.csv",
    "copper": "copper.csv",
    "oil": "oil.csv",
    "unemployment": "unemployment.csv",
    "moex": "MOEX.csv",
    "sse": "SSE.csv",
    "stoxx_600": "STOXX_600.csv"
}

# Load datasets
data = {}
for key, file in file_names.items():
    file_path = os.path.join(data_path, file)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        if "timestamp" in df.columns:
            df["timestamp"] = pd.to_datetime(df["timestamp"])
            for col in df.columns:
                if col != "timestamp":
                    df[col] = pd.to_numeric(df[col], errors='coerce')
            data[key] = df
        else:
            print(f"âš ï¸ Skipping {key}: missing 'timestamp' column")
    else:
        print(f"âŒ File not found: {file_path}")

# Rename standard columns
rename_map = {
    "sp500": {"Close": "sp500_close"},
    "copper": {"price": "copper_price"},
    "oil": {"Close": "oil_close"},
    "unemployment": {"Unemployment": "unemployment_rate"},
    "bitcoin": {"Close": "bitcoin_close"},
    "gold": {"Close": "gold_close"},
}
for key, renames in rename_map.items():
    if key in data:
        data[key] = data[key].rename(columns=renames)

# Merge datasets
sp500 = data["sp500"]
all_data = sp500[["timestamp", "sp500_close"]]

for key, df in data.items():
    if key != "sp500":
        df = df.rename(columns={col: f"{key}_{col}" for col in df.columns if col != "timestamp"})
        all_data = all_data.merge(df, on="timestamp", how="left")

# Final clean
all_data.sort_values("timestamp", inplace=True)
all_data.ffill(inplace=True)
all_data.dropna(inplace=True)

# Define prediction intervals
prediction_intervals = [7, 30, 180]
results = {}



for interval in prediction_intervals:
    all_data[f"sp500_target_{interval}"] = all_data["sp500_close"].shift(-interval)
    model_data = all_data.dropna(subset=[f"sp500_target_{interval}"]).copy()

    feature_columns = [
        col for col in model_data.columns
        if col not in ["timestamp"] + [f"sp500_target_{i}" for i in prediction_intervals]
    ]

    X = model_data[feature_columns]
    y = model_data[f"sp500_target_{interval}"]
    timestamps = model_data["timestamp"]

    X_train, X_test, y_train, y_test, timestamps_train, timestamps_test = train_test_split(
        X, y, timestamps, test_size=0.2, random_state=42
    )

    # XGBoost model
    model = XGBRegressor(n_estimators=200, max_depth=6, learning_rate=0.1, random_state=42)
    model.fit(X_train, y_train)

    # Feature importance via permutation
    importance = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
    sorted_idx = importance.importances_mean.argsort()[::-1]
    selected_features = [feature_columns[i] for i in sorted_idx if importance.importances_mean[i] > 0.01]

    print(f"\nðŸ“Š Selected features for {interval}-day prediction:", selected_features)

    # Retrain with selected features
    X_train_sel = X_train[selected_features]
    X_test_sel = X_test[selected_features]
    model.fit(X_train_sel, y_train)

    # Predict
    y_pred = model.predict(X_test_sel)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    results[interval] = {"RÂ²": r2, "MAE": mae, "MSE": mse, "RMSE": rmse}

    print(f"ðŸ“ˆ {interval}-day XGBoost forecast:")
    print(f"   RÂ²:   {r2:.4f}")
    print(f"   MAE:  {mae:.4f}")
    print(f"   MSE:  {mse:.4f}")
    print(f"   RMSE: {rmse:.4f}")

    # Plotting
    sorted_indices = np.argsort(timestamps_test.values)
    ts_sorted = timestamps_test.values[sorted_indices]
    y_test_sorted = y_test.values[sorted_indices]
    y_pred_sorted = y_pred[sorted_indices]

    plt.figure(figsize=(10, 5))
    plt.plot(ts_sorted, y_test_sorted, label="Actual", color="blue")
    plt.scatter(ts_sorted, y_pred_sorted, label="Predicted", color="red", marker="o")
    plt.title(f"S&P 500 {interval}-Day Forecast with XGBoost")
    plt.xlabel("Timestamp")
    plt.ylabel("S&P 500 Close")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Final results
print("\nâœ… Final results with XGBoost:")
for interval, metrics in results.items():
    print(f"{interval}-day â†’ RÂ²: {metrics['RÂ²']:.4f}, MAE: {metrics['MAE']:.2f}, RMSE: {metrics['RMSE']:.2f}")
```

```python
# Prepare test DataFrame
prices = pd.Series(
    data = X_test['sp500_close'].values,
    index = pd.to_datetime(timestamps_test),
    name = 'Close'
)
df2 = prices.to_frame()

feature_columns = model.get_booster().feature_names
X_test_features = X_test[feature_columns]

df2['Forecast1'] = model.predict(X_test_features).flatten()
df2['Forecast2'] = df2['Forecast1'].shift(-2)
df2['PredictedReturn2'] = df2['Forecast2'] / df2['Close'] - 1

df2 = df2.iloc[:-2].sort_index()

# Initialize portfolio
initial_money = 100000.0
cash_balance = 0.0
stock_shares = initial_money / df2['Close'].iloc[0]
next_buy_date = None

# Simulate strategy
portfolio_values = []
for date, price, pred_ret2 in zip(df2.index, df2['Close'], df2['PredictedReturn2']):
    if stock_shares > 0 and pred_ret2 <= -0.01:
        # Sell if predicted return > -1%
        cash_balance += stock_shares * price
        stock_shares = 0.0
        next_buy_date = date + pd.Timedelta(days=2)
    elif (stock_shares == 0 and next_buy_date is not None and date >= next_buy_date and pred_ret2 >= 0.01):
        # Buy if predicted return > +1%
        stock_shares = cash_balance / price
        cash_balance = cash_balance % price
        next_buy_date = None

    # Record total portfolio value
    portfolio_values.append(cash_balance + stock_shares * price)

# Attach results
df2['StrategyValue'] = portfolio_values
df2['BuyHoldValue'] = initial_capital * (df2['Close'] / df2['Close'].iloc[0])

# Compute & print returns
strategy_return = df2['StrategyValue'].iloc[-1] / initial_capital - 1
buy_hold_return = df2['BuyHoldValue'].iloc[-1] / initial_capital - 1
print(f"Strategy return: {strategy_return * 100:.2f}%, Portfolio value: ${df2['StrategyValue'].iloc[-1]:.2f}")
print(f"Buy & Hold return: {buy_hold_return * 100:.2f}%, Portfolio value: ${df2['BuyHoldValue'].iloc[-1]:.2f}")

# Plot both curves
ax = df2[['StrategyValue', 'BuyHoldValue']].plot(
    figsize = (10, 5),
    style = ['-','-'],
    linewidth = 2
)
ax.set_title('Total Portfolio Value Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Portfolio Value ($USD)')
ax.legend(['Strategy', 'Buy & Hold'])
plt.tight_layout()
plt.show()
```
